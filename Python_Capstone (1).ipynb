```python
#Task 3: Split “Name” into First and Last Name and drop the old “Name” column
# Read employee data
employee_df = pd.read_csv('employee.csv')

# Split 'Name' column into two new columns
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', n=1, expand=True)

# Remove the original 'Name' column
employee_df.drop(columns=['Name'], inplace=True)

# Save updated employee.csv
employee_df.to_csv('employee.csv', index=False)
```

```python
#Task 4: Merge all 3 DataFrames into one
# Read all three files again
employee_df = pd.read_csv('employee.csv')
seniority_df = pd.read_csv('seniority.csv')
project_df = pd.read_csv('project.csv')

# Merge employee with seniority
merged_df = employee_df.merge(seniority_df, on='ID')

# Merge with project data
final_df = merged_df.merge(project_df, on='ID')

# Save final merged data
final_df.to_csv('final.csv', index=False)
```

```python
#Task 5: Add bonus column (5% bonus for finished projects)
# Read final merged file
final_df = pd.read_csv('final.csv')

# Apply bonus logic: 5% of cost if project is finished
final_df['Bonus'] = final_df.apply(lambda row: 0.05 * row['Cost'] if row['Status'] == 'Finished' else 0, axis=1)

# Save updated final.csv
final_df.to_csv('final.csv', index=False)
```

```python
#Task 6: Demote level by 1 for failed projects and remove if level > 4
# Iterate through rows to demote designation level for failed projects
for index, row in final_df.iterrows():
    if row['Status'] == 'Failed':
        final_df.at[index, 'Designation Level'] += 1  # demote

# Drop all rows where designation level > 4
final_df = final_df[final_df['Designation Level'] <= 4]

# Save updated final.csv
final_df.to_csv('final.csv', index=False)
```

```python
#Task 7: Add "Mr." or "Mrs." to first name and drop gender
# Add title based on gender
def add_title(row):
    return 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name']

# Apply function and drop gender
final_df['First Name'] = final_df.apply(add_title, axis=1)
final_df.drop(columns=['Gender'], inplace=True)

# Save updated final.csv
final_df.to_csv('final.csv', index=False)
```

```python
#Task 8: Promote designation level by 1 for age > 29
# Apply condition to promote by reducing designation level
final_df['Designation Level'] = final_df.apply(
    lambda row: row['Designation Level'] - 1 if row['Age'] > 29 and row['Designation Level'] > 1 else row['Designation Level'],
    axis=1
)

# Save updated final.csv
final_df.to_csv('final.csv', index=False)
```

```python
#Task 9: Total project cost per employee
# Group by ID and First Name, sum up cost
total_cost_df = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# Rename column to Total Cost
total_cost_df.rename(columns={'Cost': 'Total Cost'}, inplace=True)

# Save to CSV
total_cost_df.to_csv('total_project_cost.csv', index=False)
```
